```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
### pie chart - top 3 countries that uses zomato

plt.pie(country_values[:3], labels=country_names[:3], autopct='%1.2f%%')
plt.show()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (15,8)
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings, hue='Rating color', 
            palette=['blue','red','orange','yellow','lightgreen','green'])
plt.show()
```

```python
## count plot

sns.countplot(x='Rating color', data=ratings, palette=['blue','red','orange','yellow','lightgreen','green'])
```

```python
## find top 10 cusines

cuisines_list = list(final_df['Cuisines'].apply(lambda x : str(x).split(',')))
```

```python
cuisines_list
```

```python
cuisines = []

for item in cuisines_list:
    for inner_item in item:
        cuisines.append(inner_item)
```

```python
list_of_cuisines = list(set(cuisines))
```
